```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
from math import ceil
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn import preprocessing
from sklearn.utils import shuffle

def error(y, a, b, X):
    return ((y-(a+np.dot(b,X)))[0])

def error2(y, a, b, X):
    return ((y-(a+np.dot(b,X)))[0])**2

def gd(a, b,lr, X, y, lam):
    bi=[]
    i=0
    er=error(y, a, b, X)
    for bk in b:
        bi.append(bk+X[i]*lr*(er)-lam*bk)
        i=i+1
    a=a+1*lr*(y-((a+(np.dot(b,X)))))
    return a, bi

def gdmb(a, b,lr, X, y, lam):
    bi=[]
    er=0
    i=0
    for x in X:
        er=er+error(y[i], a, b, x)/len(X)
        i=i+1
    i=0
    mean=np.mean(pd.DataFrame(X))
    for bk in b:
        bi.append(bk+mean[i]*lr*(er)-lam*bk)
        i=i+1
    a=a+1*lr*(er)
    return a, bi
```

```python
d=pd.read_csv('https://raw.githubusercontent.com/ClaseML-2017/MaterialyTareas/master/datos/regLin.csv')
```

```python
d[['X']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(pd.DataFrame(d[['X']]), pd.DataFrame(d['y']), test_size=0.33, random_state=42)


normalizer = preprocessing.StandardScaler()
normalizer.fit(X_train)
X_train = pd.DataFrame(normalizer.transform(X_train))
X_test = pd.DataFrame(normalizer.transform(X_test))

normalizer.fit(y_train)
y_train = pd.DataFrame(normalizer.transform(y_train))
y_test = pd.DataFrame(normalizer.transform(y_test))
```

```python
a2=500
b2=[500]
bs=50
lam=0.00
i=0
for j in range(1000):
    X_train, y_train=shuffle(X_train, y_train, random_state=j)
    print "\r"+str(j),
    for k in range(len(X_train)/bs):
        a2, b2=gdmb(a2, b2, 0.1, X_train.values[k*bs:(k+1)*bs], y_train.values[k*bs:(k+1)*bs], lam)
        i=i+1
te=0
for k in range(len(X_test.values)):
    te=te+error2(y_test.values[k], a2, b2,  X_test.values[k])/len(X_test.values)
    
print "\r With Lam=0.00 we got want a="+str(a2)+" and b="+str(b2)+" and error of "+str(te)
lam=0.01
i=0
for j in range(1000):
    X_train, y_train=shuffle(X_train, y_train, random_state=j)
    print "\r"+str(j),
    for k in range(len(X_train)/bs):
        a2, b2=gdmb(a2, b2, 0.1, X_train.values[k*bs:(k+1)*bs], y_train.values[k*bs:(k+1)*bs], lam)
        i=i+1
te=0
for k in range(len(X_test.values)):
    te=te+error2(y_test.values[k], a2, b2,  X_test.values[k])/len(X_test.values)
    
print "\r With Lam=0.01 we got want a="+str(a2)+" and b="+str(b2)+" and error of "+str(te)
```
